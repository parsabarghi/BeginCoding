```python
import matplotlib.pyplot as plt
```

```python
# linear chart
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [2, 34, 5, 6, 76, 8, 99, 0, 25]
# showing plot and change some info
plt.plot(x, y, color = "red")
# write a title
plt.title("my first plot")
# add labels
plt.xlabel("X")
plt.ylabel("y")
# show plot 
plt.show
```

```python
# change line to O
plt.plot(x, y, 'o', color="red")
plt.show
```

```python
# use line and o 
plt.plot(x, y, marker= 'o', color="red")
plt.show
```

```python
# make a bar chart 
x = ["red", "blue", "green", "pink"]
y = [5, 6, 11, 21]

plt.bar(x,y, color="red")
# write a title
plt.title("my first plot")
# add labels
plt.xlabel("X")
plt.ylabel("y")

plt.show
```

```python
# make a pie chart
x = ["red", "blue", "green", "pink"]
y = [5, 6, 11, 21]

plt.pie(y, labels=x, autopct = "%1.1f%%", colors=["red", "blue", "green", "pink"])
plt.show
```

```python
# make histogram
import numpy as np

x = np.random.rand(1000, 1)

plt.hist(x, bins=5)
plt.show
```

```python
# make a hitmap
x = np.random.rand(50, 50)

plt.imshow(x, cmap="hot")
plt.title("hit map")
plt.colorbar()
plt.show
```
